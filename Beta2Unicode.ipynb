```python
import re
import re
from cltk.corpus.greek.beta_to_unicode import Replacer
r = Replacer()

def addunicode(fname):
    
    f = open(fname)
    outfname = re.sub('\.xml','-unic.xml',fname)
    outf = open(outfname,'w')
    
    inbody = 0
    for l in f:
        
        l = re.sub('\s+$','',l)
        
        if(re.search('<body',l)):
            inbody = 1
            print(l,file=outf)
            continue
        if(not inbody):
            print(l,file=outf)
            continue
            
        l = re.sub('(<p>|</p>)','',l)

        l = re.sub('</seg>','',l)
        l = re.sub('<seg>','<s/>',l)
        l = re.sub('&mdash;','—',l)
        l = re.sub('&ndash;','-',l)
        l = re.sub('(&pound;|&pounds;)','£',l)
        l = re.sub('<hi>','<hi rend="italic">',l)
        l = re.sub('&auml;','ä',l)
        l = re.sub('&uuml;','ü',l)
        l = re.sub('&ouml;','ö',l)
        l = re.sub('&iuml;','ï',l)
        l = re.sub('&oelig;','oe',l)
        l = re.sub('&times;','×',l)
        l = re.sub("&sect;",'§',l)
        
            
        l = re.sub('\s*<div1 type="book" n="(1)">','\n<div type="textpart" subtype="book" n="\g<1>">',l)
        l = re.sub('<div1 type="book" n="([0-9]+)">','\n<div type="textpart" subtype="book" n="\g<1>">',l)
        l = re.sub('</div1>','</p>\n</div>\n</div>\n',l)
            
        l = re.sub('\s*<div1 type="chapter" n="(1)">','\n<div type="textpart" subtype="chapter" n="\g<1>">',l)
        l = re.sub('<div1 type="chapter" n="([0-9]+)">','</div>\n</div>\n\n<div type="textpart" subtype="chapter" n="\g<1>">',l)
            
        l = re.sub('<milestone unit="book" n="([0-9]+)"/>','<div type="textpart" subtype="section" n="\g<1>"><p>',l)
        
        
        l = re.sub('<milestone unit="chapter" n="(1)"/>','<div type="textpart" subtype="chapter" n="\g<1>">',l)
        l = re.sub('<milestone unit="chapter" n="([0-9]+)"/>','</p>\n</div>\n</div>\n<div type="textpart" subtype="chapter" n="\g<1>">',l)
        
        
        l = re.sub('<milestone unit="section" n="(1)"/>','<div type="textpart" subtype="section" n="\g<1>"><p>',l)
        l = re.sub('<milestone unit="section" n="([0-9]+)"/>','</p>\n</div>\n<div type="textpart" subtype="section" n="\g<1>"><p>',l)
            
        while(re.search('<foreign lang="greek">(.+?)</foreign>',l)):
            m = re.search('<foreign lang="greek">(.+?)</foreign>',l)
            curgreek = r.beta_code(m[1])
            print('greek',m[1],curgreek)
            l = re.sub('<foreign lang="greek">(.+?)</foreign>','<foreign xml:lang="grc">'+curgreek+'</foreign>',l,1)
        while(re.search('<quote lang="greek">(.+?)</quote>',l)):
            m = re.search('<quote lang="greek">(.+?)</quote>',l)
            curgreek = r.beta_code(m[1])
            print('quote',m[1],curgreek)
            l = re.sub('<quote lang="greek">(.+?)</quote>','<xquote xml:lang="grc">'+curgreek+'</xquote>',l,1)
        l = re.sub('<quote>','“',l)
        l = re.sub('</quote>','”',l)
        l = re.sub('xquote','quote',l)
        l = re.sub('<q>','<quote rend="blockquote">',l)
        l = re.sub('</q>','</quote>',l)
        
        print(l,file=outf)
    
    f.close()
    outf.close()
    
addunicode('smith.thuc.xml')
```
