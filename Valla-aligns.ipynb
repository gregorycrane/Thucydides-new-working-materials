```python
import re

chunkdict = {}
founddict = {}


def getchunks(fname):
    f = open(fname)
    global chunkdict
    curcit = ''
    curbook = ''
    curchap = ''
    
    
    for l in f:
        
        m = re.search('(<div[^>]+subtype="([^"]+)"[^>]*)',l)
        if(m):
            cursubtype = m[2]
            curdiv = m[1]
            n = re.search('n="([^"]+)"',l)
            if(n):
                curn = n[1]
            else:
                print('no n!',curdiv)
            
            if(cursubtype=='book'):
                curbook = curn
            if(cursubtype=='chapter'):
                curchap = curn
            if(cursubtype=='section'):
                cursection = curn
                if(not curbook or not curchap):
                    print('sect',curbook,curchap,cursection)
                curcit = curbook + '.' + curchap +'.'+ cursection
                chunkdict[curcit] = 0
                #print(curcit)
            
    
        
        l = re.sub('<[^>]+>',' ',l)
        l = re.sub('\s+$','',l)
    
        if(not l):
            continue
        if(not curcit):
            continue
        
        chunkdict[curcit] = chunkdict[curcit] + len(l.split())
    f.close()
    
    
    
fname = 'tlg0003.tlg001.perseus-grc2.xml'

getchunks(fname)


fname = 'thuc.valla1564norm_lat.xml'

f = open(fname)

for foo in chunkdict:
    founddict[foo] = 0

links = 0
for l in f:
    refs = re.findall('<ref n="([0-9]+\.[0-9]+\.[0-9]+)',l)
    links = links + len(refs)
    for foo in refs:
        if(not foo in founddict):
            print('nosect',foo)
            founddict[foo] = 0
        founddict[foo] = founddict[foo] + 1


f.close()

misses = 0
for foo in founddict:
    if(founddict[foo]==0):
        misses = misses + 1
        
print(links,misses,len(founddict),misses/len(founddict))
    
```
