```python
import re

reflist = {}
notelist = {}
notetext = {}

fname = 'betant.thuc_fr-unic.xml'

f = open(fname)

def addentry(curitem,curdict):
    if( curitem in curdict):
        curdict[curitem] = curdict[curitem] + 1
    else:
        curdict[curitem] = 1
curnote = ''
for l in f:
    l = re.sub('\s+$','',l)
    
    refs = re.findall('<note target="([^"]+)"',l)
    for ref in refs:
        addentry(ref,reflist)
        
    m = re.search('<note xml:id="([^"]+)"',l)
    if(m):
        curnote = m[1]
        addentry(curnote,notelist)
        notetext[curnote] = ''
        continue
        
    if(curnote):
        #print(curnote,l)
        if(notetext[curnote]):
            notetext[curnote] = notetext[curnote] + ' '
        notetext[curnote] = notetext[curnote] + l
        if(re.search('</note>',l)):
            curnote = ''
        
        


f.close()


for foo in reflist:
    if(not foo in notelist):
        print('missing note',foo)
        
        
for foo in notelist:
    
    if(not foo in reflist):
        print('missing ref',foo)
        
f = open(fname)
outfname = re.sub('\.xml','-note.xml',fname)
outf = open(outfname,'w')
innotes = 0
for l in f:
    l = re.sub('\s+$','',l)
    while(re.search('<note target="([^"]+)"[^>]*/>',l)):
        m = re.search('<note target="([^"]+)"[^>]*/>',l)
        curref = m[1]
        l = re.sub('<note target="([^"]+)"[^>]*/>','<note xml:lang="mul">'+notetext[curref],l,1)
    if(re.search('<div type="textpart" subtype="notes">',l)):
        innotes = 1
        continue
        
    if(innotes and re.search('</div>',l)):
        innotes = 0
        continue
    if(innotes):
        continue
    
    print(l,file=outf)

    
f.close()
outf.close()
        
        
#print(len(reflist),len(notelist))   
```
