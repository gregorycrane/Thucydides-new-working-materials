```python
import re

fname = 'tlg0003.tlg001.1st1K-eng2-nosecs.xml'
workfname = '/tmp/tlg0003.tlg001.1st1K-eng2-nosecs.xml'

f = open(fname)
outf = open(workfname,'w')
for l in f:
    l = re.sub('\s*(<milestone[^>]+>)\s*','\n\g<1>\n',l)
    l = re.sub('\s*(<quote[^>]*>|</quote>)\s*','\n\g<1>\n',l)
    l = re.sub('\s*(<note[^>]*>|</note>)','\n\g<1>\n',l)
    
    print(l,file=outf)

f.close()
outf.close()

f = open(workfname)
outf = open('tlg0003.tlg001.1st1K-eng2.xml','w')
inquote = 0
innote = 0
    

for l in f:
    if(re.search('<note',l)):
        innote = 1
        
    if(re.search('</note>',l)):
        innote = 0
        
    if(not innote):
        
        if(re.search('<quote',l)):
            inquote = 1
        
        if(re.search('</quote>',l)):
            inquote = 0
        
    m = re.search('(</p> </div> </div> <div type="textpart" subtype="chapter" n="[0-9]">)',l)
    if(m and inquote):
        l = re.sub('(</p>) (</div> </div> <div type="textpart" subtype="chapter" n="[0-9]">)','</quote> \g<1> \g<2>',l)
        
        
    m = re.search('(<div type="textpart" subtype="section" n="1"> <p>)',l)
    if(m):
        if(inquote):
            l = re.sub('(<div type="textpart" subtype="section" n="1">) <p>','\g<1><p><quote rend="merge">',l)
        
    m = re.search('<milestone unit="section" n="([0-9]+)"/>',l)
    
    if(m):
        if(inquote):
            l = re.sub('<milestone unit="section" n="([0-9]+)"/>','</quote>\n</p>\n</div>\n<div type="textpart" subtype="section" n="\g<1>"><p><quote rend="merge">',l)
    
        else:
            l = re.sub('<milestone unit="section" n="([0-9]+)"/>','\n</p>\n</div>\n<div type="textpart" subtype="section" n="\g<1>"><p>',l)

    print(l,file=outf)
    print(innote,inquote,l)

f.close()
outf.close()
```
